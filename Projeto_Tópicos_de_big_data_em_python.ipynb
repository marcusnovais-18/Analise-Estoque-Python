## 4. Limpeza e Prepara√ß√£o dos Dados
Dados "sujos" (como " ATIVO" com espa√ßo) geram an√°lises incorretas.
* Removemos espa√ßos em branco das colunas de texto.
* Padronizamos a coluna `STATUS` para ficar sempre em mai√∫sculo (ex: "Inativo" vira "INATIVO").

```python
# Limpeza e Prepara√ß√£o dos Dados
# ---
# Colunas de texto que usaremos nas an√°lises
colunas_para_limpar = ['CATEGORIA', 'STATUS', 'Tipo Produto']

print("Iniciando limpeza dos dados...")
for col in colunas_para_limpar:
    # Verifica se a coluna realmente existe no DataFrame
    if col in df.columns:

        # .astype(str) for√ßa a coluna a ser do tipo string (texto)
        # .str.strip() remove espa√ßos em branco do in√≠cio e do fim
        # Ex: "  ATIVO  " vira "ATIVO"
        df[col] = df[col].astype(str).str.strip()

        # Padroniza a coluna 'STATUS' para mai√∫sculas
        # Ex: "Inativo" vira "INATIVO"
        if col == 'STATUS':
            df[col] = df[col].str.upper()

        print(f"Coluna '{col}' limpa e padronizada.")
    else:
        # Aviso caso uma coluna esperada n√£o seja encontrada
        print(f"Aten√ß√£o: Coluna '{col}' n√£o encontrada no DataFrame.")

print("\nLimpeza de dados conclu√≠da.")
```

Aqui, filtramos a base para encontrar apenas os produtos `INATIVO` e contamos quais categorias aparecem mais.

```python
# An√°lise 1 (Processamento) - Produtos INATIVOS

# 1. Filtra o DataFrame original para criar um novo (df_inativos)

# Contendo apenas linhas onde a coluna 'STATUS' √© igual a 'INATIVO'

df_inativos = df[df['STATUS'] == 'INATIVO'].copy()

# 2. Conta a ocorr√™ncia de cada 'CATEGORIA' dentro do DataFrame de inativos

# .value_counts() j√° ordena do maior para o menor

contagem_inativos = df_inativos['CATEGORIA'].value_counts()

# 3. Exibe o Top 10 categorias com mais itens inativos

print("--- Top 10 Categorias com Mais Produtos INATIVOS ---")

print(contagem_inativos.head(10))
```

## Engenharia de Atributos e Modelagem Preditiva

Como a base de dados original ("Planilha Insumos") √© composta majoritariamente por dados categ√≥ricos (Texto), aplicamos t√©cnicas de **Engenharia de Atributos** para extrair valores num√©ricos que permitam a realiza√ß√£o de testes estat√≠sticos rigorosos.

**Novas Vari√°veis Criadas para An√°lise:**
1.  **`status_bin`:** Vari√°vel alvo codificada (1 = ATIVO, 0 = INATIVO).
2.  **`fiscal_bin`:** Vari√°vel preditora codificada (1 = SIM, 0 = NAO).
3.  **`tamanho_descricao`:** Contagem de caracteres do nome do produto. Serve como proxy para "Qualidade do Cadastro" (descri√ß√µes muito curtas podem indicar itens antigos ou mal cadastrados).

```python
# --- ENGENHARIA DE ATRIBUTOS (Transformando Texto em N√∫meros) ---

# 1. Codifica√ß√£o Bin√°ria (0 ou 1)
# Se STATUS for 'ATIVO', vira 1. Caso contr√°rio, 0.
df['status_bin'] = np.where(df['STATUS'].astype(str).str.strip().str.upper() == 'ATIVO', 1, 0)

# Se USA FISCAL for 'SIM' ou 'S', vira 1.
df['fiscal_bin'] = np.where(df['USA FISCAL'].astype(str).str.strip().str.upper().isin(['SIM', 'S']), 1, 0)

# 2. Cria√ß√£o de Vari√°vel Cont√≠nua (Tamanho da Descri√ß√£o)
# Calcula quantos caracteres tem a descri√ß√£o do produto
df['tamanho_descricao'] = df['DESCRICAO'].astype(str).str.len()

# --- AN√ÅLISE DESCRITIVA ---
print("--- Estat√≠sticas Descritivas das Novas Vari√°veis ---")
display(df[['status_bin', 'fiscal_bin', 'tamanho_descricao']].describe())
```

O Heatmap (mapa de calor) exibe os coeficientes de correla√ß√£o de Pearson entre as vari√°veis num√©ricas e bin√°rias transformadas:

* **Aus√™ncia de Correla√ß√£o Forte:** A predomin√¢ncia de cores claras (tons de cinza/azul p√°lido) e valores pr√≥ximos de zero indica que **n√£o existe uma correla√ß√£o linear forte** entre as vari√°veis analisadas.

Sendo negativa, confirma estatisticamente o que vimos nos gr√°ficos anteriores: existe uma leve tend√™ncia de que, quanto **maior** a descri√ß√£o, **menor** a chance de o produto ser ativo (ou seja, produtos ativos tendem a ter descri√ß√µes mais curtas).

#4. Regress√£o Linear (Linear Probability Model)
Para finalizar, aplicamos um modelo de **Regress√£o Linear (OLS)**.
Como nossa vari√°vel alvo (`status_bin`) √© bin√°ria (0 ou 1), este modelo estima a **probabilidade** de um item ser ativo com base em suas caracter√≠sticas.

**Modelo:** `status_bin ~ fiscal_bin + tamanho_descricao`

```python
# --- MODELO DE REGRESS√ÉO LINEAR (Probabilidade de Atividade) ---

# √â NECESS√ÅRIO IMPORTAR A BIBLIOTECA PARA USAR O 'smf'
import statsmodels.formula.api as smf
import pandas as pd # Recomenda-se importar pandas tamb√©m, caso o DF n√£o esteja carregado

print("--- MODELO DE REGRESS√ÉO LINEAR (Probabilidade de Atividade) ---")

# Criando o modelo: Prever STATUS com base em FISCAL e TAMANHO DA DESCRI√á√ÉO
# OLS = Ordinary Least Squares (M√≠nimos Quadrados Ordin√°rios)
modelo = smf.ols('status_bin ~ fiscal_bin + tamanho_descricao', data=df).fit()

print("\n--- Sum√°rio do Modelo ---")
print(modelo.summary())
```
